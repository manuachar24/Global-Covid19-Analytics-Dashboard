```python
df.drop(['NewCases','NewDeaths','NewRecovered','WHO Region', 'iso_alpha','Serious,Critical'],axis=1,inplace=True)
```

```python
df.columns=df.columns.str.lower()
# df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'Country/Region': 'country'}, inplace=True)
```

```python
df['deathrate'] = (df['totaldeaths'] / df['totalcases']) * 100
df['recoveryRate'] = (df['totalrecovered'] / df['totalcases']) * 100
df['activeRate'] = (df['activecases'] / df['totalcases']) * 100
```

```python
df.columns
```

```python
df['totaldeaths'].fillna(df['totaldeaths'].mean(),inplace=True)
```

```python
df['deaths/1m pop'].fillna(df['deaths/1m pop'].mean(),inplace=True)
```

```python
df['totaltests'].fillna(df['totaltests'].mean(),inplace=True)
```

```python
df['tests/1m pop'].fillna(df['tests/1m pop'].mean(),inplace=True)
```

```python
df['population'].fillna(df['population'].mean(),inplace=True)
```

```python
df['totalrecovered'].fillna(df['totalrecovered'].median(),inplace=True)
```

```python
df['activecases'].fillna(df['activecases'].median(),inplace=True)
```

```python
df['tot cases/1m pop'].fillna(df['tot cases/1m pop'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df["continent"].fillna("Asia",inplace=True)
```

```python
df.isnull().sum()
```

```python
df['continent'].head(10)
```

```python
top_cases = df.sort_values(by='totalcases', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_cases['country/region'], top_cases['totalcases'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Total COVID Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.show()
```

```python
continent_cases = df.groupby('continent')['totalcases'].sum()

plt.figure(figsize=(6,6))
plt.pie(continent_cases, labels=continent_cases.index, autopct='%1.1f%%')
plt.title('Total COVID Cases by Continent')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['totaltests'], df['totalcases'])
plt.title('Total Tests vs Total Cases')
plt.xlabel('Total Tests')
plt.ylabel('Total Cases')
plt.show()
```

```python
top_recovered = df.sort_values(by='totalrecovered', ascending=False).head(10)

plt.figure(figsize=(10,4))
plt.scatter(top_recovered['country/region'], top_recovered['totalrecovered'])
plt.title('Top 10 Countries by Total Recovered Cases')
plt.xlabel('Recovered Cases')
plt.ylabel('Country')
plt.show()
```

```python
import seaborn as sns

numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
top_active = df.sort_values(by='activecases', ascending=False).head(10)

plt.figure(figsize=(6,4))
plt.bar(top_active['country/region'], top_active['activecases'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Active Cases')
plt.xlabel('Country')
plt.ylabel('Active Cases')
plt.show()
```
